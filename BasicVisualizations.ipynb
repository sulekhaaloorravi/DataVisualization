### Line plot

```python
x = [1990,1991,1992,1993,1994,1995,1996,1997,1998]
```

```python
y = [1,2,3,4,5,6,7,8,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
y1 = np.sin(y)
```

```python
plt.plot(x,y1)
plt.show()
```

### Bar Chart

```python
plt.bar(x,y)
plt.show()
```

### Histogram Plot

```python
from numpy.random import randn
from numpy.random import seed
```

```python
seed(123)
```

```python
data = randn(100)
```

```python
plt.hist(data)
plt.show()
```

### Box and Whisker Plot

```python
boxdata = [randn(100), 3 * randn(100), 9 * randn(100)]
```

```python
plt.boxplot(boxdata)
plt.show()
```
